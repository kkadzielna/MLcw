```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import tensorflow
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_openml

from tensorflow.keras.datasets import fashion_mnist
```

```python
mnist = fetch_openml(name='Fashion-MNIST')
X = np.array(mnist.data)
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(mnist.data, mnist.target,test_size=0.1, random_state=40)
print(X_train.shape); print(X_test.shape)
print(y_train.shape); print(y_test.shape)
```

```python
#scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#shuffle
np.random.seed(42) 
shuffle_index = np.random.permutation(63000)
X_train, y_train = X_train[shuffle_index], y_train[shuffle_index]
np.random.seed(42)
shuffle_index_test = np.random.permutation(7000)
X_test, y_test = X_test[shuffle_index_test], y_test[shuffle_index_test]

#reduce the dataset
X_train_reduced, y_train_reduced = X_train[0:1000], y_train[0:1000]
X_test_reduced, y_test_reduced = X_test[0:200], y_test[0:200]
```

```python
from sklearn.neural_network import MLPClassifier

#an untuned classifier
#mlp = MLPClassifier(hidden_layer_sizes=(10, 10, 10), max_iter=1000, validation_fraction=0.2, early_stopping=True)
mlp = MLPClassifier(random_state = 40)
mlp.fit(X_train, y_train)
plt.plot(mlp.loss_curve_)
plt.title("Loss Curve", fontsize=14)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.show()

#warm_start
mlp = MLPClassifier(early_stopping = True, warm_start = True, random_state = 40)
mlp.fit(X_train, y_train)

print (mlp.score(X_train,y_train))
print (mlp.score(X_test,y_test))
plt.plot(mlp.loss_curve_)
plt.title("Loss Curve", fontsize=14)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

#

param_grid = {
    'hidden_layer_sizes': [(784,), (392,), (196,), (83,), (40,)],
    'activation': ['identity', 'logistic','tanh', 'relu'],
    'alpha': np.logspace(-3, -0, 10), #np.arange(0.0, 0.01, 0.0001),
    'learning_rate_init': np.arange(0.0, 0.01, 0.001),
    'max_iter' : [50, 100, 200, 300, 500],
}


tuned_mlp = GridSearchCV(mlp, param_grid, cv = 5, scoring='accuracy', verbose = 2, n_jobs=-1)
# perform search
tuned_mlp.fit(X_train, y_train)

# best estimator
params = tuned_mlp.best_params_

print(params)
print(tuned_mlp.score(X_train,y_train))
print(tuned_mlp.score(X_test,y_test))
```

```python
from sklearn.model_selection import learning_curve

#plot learning curve function from the labs
def plot_learning_curve(n, mlp, xs, ys, title):
    train_sizes, train_scores, valid_scores = learning_curve(estimator = mlp, X = xs, y = ys, cv = n)

    train_mean = np.mean(train_scores, axis=1)
    valid_mean = np.mean(valid_scores, axis=1)

    plt.plot(train_sizes, train_mean, color='blue', label='Training Accuracy')
    plt.plot(train_sizes, valid_mean, color='green', label='Validation Accuracy')

    plt.xlabel('Dataset size')
    plt.ylabel('Accuracy')
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
from sklearn.model_selection import validation_curve

def plot_validation_curve(mlp, X, y, p_name, p_range, title):
    train_scores, test_scores = validation_curve(estimator=mlp,
                                             X=X, y=y,
                                             cv=5,
    param_name=p_name, param_range=p_range)
    train_mean = np.mean(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)

    plt.plot(p_range, train_mean,
             color='blue', label='Training Accuracy')
    plt.plot(p_range, test_mean,
             color='green', label='Validation Accuracy')
    plt.xlabel(p_name)
    plt.ylabel('Accuracy')
    plt.title(title)
    plt.grid()
    plt.show()
```

```python
#on a reduced dataset due to long time
layer_range = [(784,), (392,), (196,), (83,), (40,)]
plot_validation_curve(mlp_t, X_train_reduced, y_train_reduced, 'hidden_layer_sizes', layer_range, 'figure 6.')

activation_range = list(['identity', 'logistic','tanh', 'relu'])
plot_validation_curve(mlp_t, X_train_reduced, y_train_reduced, 'activation', activation_range, 'figure 7.')

alpha_range = list(np.linspace(0, 10, 100))#list(np.linspace(0, 10, 1000))
plot_validation_curve(mlp_t, X_train_reduced, y_train_reduced, 'alpha', alpha_range, 'figure 8.')

lr_range = list(np.arange(0.0, 0.01, 0.001))
plot_validation_curve(mlp_t, X_train_reduced, y_train_reduced, 'learning_rate_init', lr_range, 'figure 9.')

max_iter_range = list([50, 100, 200, 300, 500])
plot_validation_curve(mlp_t, X_train_reduced, y_train_reduced, 'max_iter', max_iter_range, 'figure 10.')
```

```python
from matplotlib.colors import ListedColormap
def plot_decision_boundaries(X, y, model, title):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    colors = {'0':'red', '1':'green', '2':'blue', '3':'yellow', '4':'orange', '5':'pink', '6':'navy', '7':'grey', '8':'purple', '9':'violet'}  
    print(Z.shape)
    print(Z.ravel().shape)
    print(X[:, 0].shape)
    print(X[:, 1].shape)
    
    colored_labels = np.array([colors[xi] for xi in y]) #Z.ravel()
    plt.contourf(xx, yy, Z, cmap='viridis')


    cb = plt.colorbar()
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.title(title)
    plt.scatter(X[:, 0], X[:, 1], c=colored_labels, cmap='viridis', s=10)

    plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_train)
mlp_t.fit(np.c_[X_pca[:,0], X_pca[:,1]], y_train)
plot_decision_boundaries(X_pca, y_train, mlp_t, 'figure 12.')
```

```python
from sklearn.model_selection import learning_curve

#plot learning curve function from the labs
def plot_fit_times(n, clf1, clf2, xs, ys, title):
    train_sizes1, train_scores1, valid_scores1, fit_times1,_ = learning_curve(estimator = clf1, X = xs, y = ys, cv = n, return_times = True)
    train_sizes2, train_scores2, valid_scores2, fit_times2,_ = learning_curve(estimator = clf2, X = xs, y = ys, cv = n, return_times = True)

    fit_times_mean1 = np.mean(fit_times1, axis=1)
    fit_times_mean2 = np.mean(fit_times2, axis=1)

    plt.plot(train_sizes1, fit_times_mean1, color='red', label='mlp')
    plt.plot(train_sizes2, fit_times_mean2, color='green', label='svm')   
    plt.xlabel('Dataset size')
    plt.ylabel('Time')
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
classifiers = ["mlp", 'svm']
times = [mlp_time_smol, svm_time_smol]
plt.bar(classifiers,times)
plt.xlabel('training methods')
plt.ylabel("times")
plt.title('figure 20.')
plt.show()

classifiers = ["mlp", "mlp - test", "svm", "svm - test"]
scores = [mlp_t.score(X_train_reduced,y_train_reduced), mlp_t.score(X_test_reduced,y_test_reduced), svm_t_smol.score(X_train_reduced,y_train_reduced), svm_t_smol.score(X_test_reduced,y_test_reduced)]
plt.bar(classifiers,scores)
plt.xlabel('training methods')
plt.ylabel("accuracy")
plt.title('figure 21.')
plt.show()
```

```python
def plot_learning_curves_comparison(n, clf1, clf2, xs, ys, title):
    train_sizes1, train_scores1, valid_scores1 = learning_curve(estimator = clf1, X = xs, y = ys, cv = n)

    train_mean1 = np.mean(train_scores1, axis=1)
    valid_mean1 = np.mean(valid_scores1, axis=1)

    plt.plot(train_sizes1, train_mean1, color='blue', label='Training Accuracy - MLP')
    plt.plot(train_sizes1, valid_mean1, color='green', label='Validation Accuracy - MLP')
    
    train_sizes2, train_scores2, valid_scores2 = learning_curve(estimator = clf2, X = xs, y = ys, cv = n)

    train_mean2 = np.mean(train_scores2, axis=1)
    valid_mean2 = np.mean(valid_scores2, axis=1)

    plt.plot(train_sizes2, train_mean2, color='red', label='Training Accuracy - SVM')
    plt.plot(train_sizes2, valid_mean2, color='orange', label='Validation Accuracy - SVM')
    plt.xlabel('Dataset size')
    plt.ylabel('Accuracy')
    plt.title(title)
    plt.legend()
    plt.show()
    
mlp_t_smol = MLPClassifier(early_stopping = True, warm_start = False, random_state = 40, activation = 'tanh', alpha = 1.281, hidden_layer_sizes = (83,), learning_rate_init = 0.003, max_iter = 50)
mlp_t_smol.fit(X_train_reduced, y_train_reduced)
plot_learning_curves_comparison(5, mlp_t_smol, svm_t_smol, X_train[0:12000], y_train[0:12000], 'figure 22.')
```
